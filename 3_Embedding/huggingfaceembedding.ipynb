###Sentense Transformers on Hugging Face

Hugging face sentense transformers is a python framework for state-of-the-art sentence, text and image embeddings. 
One of the embedding models is used in the HuggingFaceEmbeddings class. We hgave also added an alias for SentenceTransformerEmbeddings for user who are more familiar with directly using this package.

```python
##%pip install langchain-huggingface
##%pip install sentence-transformers
#%pip install hf_xet
from langchain_huggingface import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
test="this is a test documents"
query_result=embeddings.embed_query(test)
query_result
```

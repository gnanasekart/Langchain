```python
#%pip install langchain-openai

import os
from dotenv import load_dotenv
load_dotenv() # Load environment variables from .env file
```

```python
os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY") # Ensure the OpenAI API key is set from the environment variables

from langchain_openai import OpenAIEmbeddings
embeddings=OpenAIEmbeddings(model="text-embedding-3-large")
embeddings
```

```python
text="Hello world"
query_result=embeddings.embed_query(text)  # Get the embedding for the text
query_result
```

```python
len(query_result)  # Check the length of the embedding vector, Defualt dimension is 3072
```

```python
##custom dimension

from langchain_community.document_loaders import TextLoader

loader=TextLoader("speech.txt")
doc=loader.load()

from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
final_docs = text_splitter.split_documents(doc)
final_docs
```

```python
#%pip install chromadb
embeddings=OpenAIEmbeddings(model="text-embedding-3-large", dimensions=1024)  # Specify a custom dimension

#vector embedding and vector storeDB
from langchain_community.vectorstores import Chroma

db=Chroma.from_documents(final_docs, embeddings)
db
```

```python
##retrieve the results from query vectorstore db
query="We have borne with their present government through all these"
retrieved_results=db.similarity_search(query)
print(retrieved_results[0].page_content)  # Print the content of the first retrieved document
```
